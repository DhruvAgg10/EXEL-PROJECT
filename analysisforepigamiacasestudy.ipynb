```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_excel('/content/Epigamia x Mayhem - Round 2.xlsx')
```

```python
dataset.head()
```

```python
# Drop row 1 (second row)
dataset = dataset.drop(index=0).reset_index(drop=True)
# Set row 0 as header
dataset.columns = dataset.iloc[0]   # Assign first row as column names
dataset = dataset[1:].reset_index(drop=True)  # Remove the first row and reset index
```

```python
dataset.head()
```

```python
# Drop column at index 1
dataset = dataset.drop(columns=dataset.columns[0])
```

```python
dataset.head()
```

```python
dataset['expected sales'] = dataset['Units Sold'] * dataset['Average Price']
dataset['Revenue loss']= dataset['expected sales'] - dataset['Sales']
```

```python

dataset.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Plot the repeat rate as grouped bar chart
plt.figure(figsize=(12, 6))
# Pivot table: Product Category vs Channel with Repeat Rate
pivot_df = dataset.pivot_table(index='Product Category', columns='Channel', values='Repeat Rate', aggfunc='mean')

pivot_df.plot(kind='bar', width=0.7, edgecolor='black', colormap='Set2')

plt.title("Repeat Rate of Product Categories by Channel")
plt.ylabel("Repeat Rate")
plt.xlabel("Product Category")
plt.xticks(rotation=45)
plt.legend(title="Channel")
plt.tight_layout()
plt.show()
```

```python
# Create pivot table: Sales by Product Category and Channel
sales_pivot = dataset.pivot_table(index='Product Category', columns='Channel', values='Sales', aggfunc='sum')

# Plot
plt.figure(figsize=(12, 6))
sales_pivot.plot(kind='bar', width=0.7, edgecolor='black', colormap='Paired')

plt.title("Sales of Product Categories by Channel")
plt.ylabel("Total Sales")
plt.xlabel("Product Category")
plt.xticks(rotation=45)
plt.legend(title="Channel")
plt.tight_layout()
plt.show()
```

```python
# Ensure months follow a correct order
month_order = ['Jan', 'Feb', 'Mar', 'Oct', 'Nov', 'Dec']
dataset['Months'] = pd.Categorical(dataset['Months'], categories=month_order, ordered=True)

# Define product groups
milk_products = ['Milk Based Drink A', 'Milk Based Drink B', 'Milk Based Drink C']
yogurt_products = ['Yogurt A', 'Yogurt B', 'Yogurt C']

# Define colors for each product
colors_milk = {'Milk Based Drink A': 'blue', 'Milk Based Drink B': 'green', 'Milk Based Drink C': 'red'}
colors_yogurt = {'Yogurt A': 'purple', 'Yogurt B': 'orange', 'Yogurt C': 'cyan'}

# Function to plot grouped products
def plot_grouped_products(product_list, title, color_dict):
    plt.figure(figsize=(10, 6))

    for product in product_list:
        product_df = dataset[dataset['Product Category'] == product]
        pivot_df = product_df.pivot_table(index='Months', columns='Region', values='Sales', aggfunc='sum')

        # Plot each product with its unique color
        for region in pivot_df.columns:
            plt.plot(pivot_df.index, pivot_df[region], marker='o', linestyle='-', color=color_dict[product], label=f"{product} - {region}")

    plt.title(title)
    plt.xlabel("Month")
    plt.ylabel("Sales")
    plt.xticks(rotation=45)
    plt.legend(title="Product & Region", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot for Milk Products
plot_grouped_products(milk_products, "Monthly Sales of Milk Products in Different Regions", colors_milk)

# Plot for Yogurt Products
plot_grouped_products(yogurt_products, "Monthly Sales of Yogurt Products in Different Regions", colors_yogurt)
```

```python
# Create pivot: Revenue loss by Product and Region
pivot_loss = dataset.pivot_table(
    index='Product Category',
    columns='Region',
    values='Revenue loss',
    aggfunc='sum'
)

# Plot
pivot_loss.plot(kind='bar', figsize=(12, 6), colormap='Set3', edgecolor='black')
plt.title('Revenue Loss by Product Category Across Regions')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue Loss ()')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Region')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
